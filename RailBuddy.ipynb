```python
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from langgraph.graph import START, StateGraph
from typing_extensions import List, TypedDict


# 1. Convert your string prompt to a PromptTemplate
prompt_template = PromptTemplate.from_template('''
You are RailBuddy an assistant for question-answering tasks for train schedules and Information.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know. Use five sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:''')

# Define state for application
class State(TypedDict):
    question: str
    context: List[Document]
    answer: str


# Define application steps
def retrieve(state: State):
    retrieved_docs = vector_store.similarity_search(state["question"])
    return {"context": retrieved_docs}
# RAG pipeline
def generate(state: State):
    docs_content = "\n\n".join(doc.page_content for doc in state["context"])
    # Use the prompt template with proper parameters
    messages = prompt_template.invoke({"question": state["question"], "context": docs_content})
    response = llm.invoke(messages)
    return {"answer": response.content}

# 3. Create your chain properly
chain = prompt_template | llm | StrOutputParser()
# Build the graph correctly
graph_builder = StateGraph(State)
graph_builder.add_node("retrieve", retrieve)
graph_builder.add_node("generate", generate)
graph_builder.set_entry_point("retrieve")
graph_builder.add_edge("retrieve", "generate")
graph = graph_builder.compile()
```

```python
# 1. Test the graph workflow
train_query = "can you tell your name what is your job and tell me how to reach Shebin El-Kom from tanta?"
graph_result = graph.invoke({"question": train_query})
print("Graph Workflow Answer:")
print(graph_result["answer"])
```
